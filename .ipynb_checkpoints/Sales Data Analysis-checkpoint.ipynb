## Import the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## Read data using read_csv function

```python
d1=pd.read_csv("salesData2010.csv",header=0,skip_blank_lines=True,names=['Day','SalesA','SalesB','SalesC','SalesD'])
```

```python
d2=pd.read_csv("salesData2011.csv",header=0,skip_blank_lines=True,names=['Day','SalesA','SalesB','SalesC','SalesD'])
```

check the datafarmes

```python
d1.head()
```

```python
d2.head()
```

## Merge both file to create a single data frame

```python
mrg=pd.merge(d1,d2,how="outer",on=None,left_on=None,right_on=None)
```

```python
mrg.head()
```

Check the type of the Day column

```python
type(mrg['Day'][0])
```

Convert the column type from string to datetime format in Pandas dataframe

```python
mrg["Day"]=pd.to_datetime(mrg.Day,format="%Y-%m-%d")
```

```python
type(mrg['Day'][0])
```

## Mean of sales for every store year wise

```python
d1="2010-01-01"
d2="2010-12-31"
d3="2011-01-01"
d4="2011-12-31"
date1=pd.to_datetime(d1,format="%Y-%m-%d")
date2=pd.to_datetime(d2,format="%Y-%m-%d")
date3=pd.to_datetime(d3,format="%Y-%m-%d")
date4=pd.to_datetime(d4,format="%Y-%m-%d")
```

```python
year1=mrg[(date1<=mrg.Day)&(mrg.Day<=date2)]
mean1=year1.mean()
year2=mrg[(date3<=mrg.Day)&(mrg.Day<=date4)]
mean2=year2.mean()         
```

```python
print("Year Wise Mean")
print("\n")
print("Mean of 2010")
print(mean1)
print("\n")
print("Mean of 2011")
print(mean2)
```

## Line chart of year wise mean sales of all store

```python
plt.figure(figsize=(10,6))
x=[2010,2011]
y1=[mean1.SalesA,mean2.SalesA]
y2=[mean1.SalesB,mean2.SalesB]
y3=[mean1.SalesC,mean2.SalesC]
y4=[mean1.SalesD,mean2.SalesD]
plt.xlabel("Year")
plt.ylabel("Mean Sales")
plt.title("Year Wise Mean")
plt.plot(x,y1,linewidth=2.0,color="g",label="SalesA")
plt.plot(x,y2,linewidth=2.0,color="r",label="SalesB")
plt.plot(x,y3,linewidth=2.0,color="y",label="SalesC")
plt.plot(x,y4,linewidth=2.0,color="m",label="SalesD")
plt.grid(True)
plt.legend(loc="upper right",title="Label",fancybox=True)
plt.show()
```

## Mean of sales for every store month wise

```python
month=pd.date_range("2009-12-31","2011-12-31",freq="M")
```

```python
month
```

```python
i=0
m1=list()
m2=list()
m3=list()
m4=list()
print("month Wise Mean")
print("\n")
while i<(len(month)-1):
    mon=mrg[(month[i]<=mrg.Day)&(mrg.Day<=month[i+1])]
    monthmean=mon.mean()
    m1.append(monthmean.SalesA)
    m2.append(monthmean.SalesB)
    m3.append(monthmean.SalesC)
    m4.append(monthmean.SalesD)
    print(monthmean)
    print(month[i+1])
    print("\n")
    i+=1
```

## Line chart of month wise mean sales of all store

```python
plt.figure(figsize=(10,6))
x=pd.date_range("2010-01-01","2011-12-31",freq="M")
y1=m1
y2=m2
y3=m3
y4=m4
plt.xlabel("Month")
plt.ylabel("Mean Sales")
plt.title("Month Wise Mean")
plt.plot(x,y1,linewidth=2.0,color="g",label="SalesA")
plt.plot(x,y2,linewidth=2.0,color="r",label="SalesB")
plt.plot(x,y3,linewidth=2.0,color="y",label="SalesC")
plt.plot(x,y4,linewidth=2.0,color="m",label="SalesD")
plt.grid(True)
plt.legend(loc="upper left",fancybox=True,title="Legend")
plt.show()
```
